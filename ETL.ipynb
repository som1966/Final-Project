```python
UN_Unemployment_Rate_2020.drop(columns=['Series','Footnotes','Source'], axis=1, inplace=True)
```

```python
UN_Unemployment_Rate_2020_rename = UN_Unemployment_Rate_2020.rename(columns={"Value": "Unemployment Rate-Value"})
UN_Unemployment_Rate_2020_rename
```

```python
frames = [UN_Life_Expectancy_2020_rename, UN_Per_Capita_GDP_2019_rename, UN_Population_Density_2019_rename, UN_Unemployment_Rate_2020_rename]
clusteredtables4to7_df = pd.concat(frames, axis=1).reindex(UN_Country_Table.index)
clusteredtables4to7_df
```

```python
# What data is missing?
# find null values
for column in clusteredtables4to7_df.columns:
    print(f"Column{column} has {clusteredtables4to7_df[column].isnull().sum()} null values")
```

```python
# Drop null rows
clusteredtables4to7_df = clusteredtables4to7_df.dropna()
clusteredtables4to7_df.head(11)
```

```python
clusteredtables4to7_df.drop(columns=['country_Name','year'], axis=1, inplace=True)
clusteredtables4to7_df
```

```python
column_names = ["Country_Name", "Year","Both sexes' Life expectancy-Value", "Per Capita GDP-Value", "Population Density-Value", "Unemployment Rate-Value" ]

clusteredtables4to7_df = clusteredtables4to7_df.reindex(columns=column_names)
clusteredtables4to7_df
```
