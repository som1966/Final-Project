## Modeling -- linear regression using all 8 indep variables except "country_name", "country_code",  "life_expectancy" & "consumer_price_index"

```python
# select features 
target = "life_ladder"
features = happiness_joined_df.columns.drop([target,"country_code","country_name", "life_expectancy", "consumer_price_index"])
X = happiness_joined_df[features]
y = happiness_joined_df[target]
import statsmodels.api as sm
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
# Normal train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78, train_size=0.92)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
# instantiate model
model = linear_model.LinearRegression()
# fit model
model.fit(X, y)

# fit
model.fit(X_train, y_train)
# predict
y_pred = model.predict(X_test)
from sklearn.preprocessing import StandardScaler
# 3) Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# 4) Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print(np.mean(X_train_scaled[:,0]))
print(np.mean(X_test_scaled[:,0]))
print(np.std(X_train_scaled[:,0]))
print(np.std(X_test_scaled[:,0]))
```

```python
# model evaluation
# ref: https://towardsdatascience.com/multiple-regression-as-a-machine-learning-algorithm-a98a6b9f307b
from sklearn.metrics import r2_score, mean_squared_error
score = r2_score(y_test, y_pred)
print("R2score:", score)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print('root mean squared error: ', rmse)
# ref: https://datagy.io/python-sklearn-linear-regression/
```

```python
# Make predictions using the test data
y_pred = model.predict(X_test)
results = pd.DataFrame({
    "Prediction": y_pred, 
    "Actual": y_test
}).reset_index(drop=True)
results.head()
```
